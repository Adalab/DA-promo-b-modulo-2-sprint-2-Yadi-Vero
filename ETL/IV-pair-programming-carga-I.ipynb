```python
def crear_bbdd(nombre_bbdd):

    mydb = mysql.connector.connect(
      host="localhost",
      user="root",
      password="AlumnaAdalab" 
    )
    print("Conexi√≥n realizada con √©xito")
    
    mycursor = mydb.cursor()

    try:
        mycursor.execute(f"CREATE DATABASE IF NOT EXISTS {nombre_bbdd};")
        print(mycursor)
    except mysql.connector.Error as err:
        print(err)
        print("Error Code:", err.errno)
        print("SQLSTATE", err.sqlstate)
        print("Message", err.msg)
```

*CREAMOS LAS TABLAS*

```python
def crear_tabla(nombre_bbdd, query):
    
    cnx = mysql.connector.connect(user='root', password="AlumnaAdalab",
                                     host='127.0.0.1', database=f"{nombre_bbdd}")
    mycursor = cnx.cursor()
    
    try: 
        mycursor.execute(query)
        cnx.commit() 
    except mysql.connector.Error as err:
        print(err)
        print("Error Code:", err.errno)
        print("SQLSTATE", err.sqlstate)
        print("Message", err.msg)
```

```python
tabla_ataques= '''
CREATE TABLE IF NOT EXISTS `tiburones`.`ataques` (
  `id` INT NOT NULL AUTO_INCREMENT,
  `id_paises` DOUBLE NOT NULL,
  `year` DOUBLE NOT NULL, 
  `age` DOUBLE NOT NULL,
  `latitud_x` DOUBLE NOT NULL,
  `longitud_x` DOUBLE NOT NULL,
  `fatal_N` DOUBLE NOT NULL,
  PRIMARY KEY (`id`))
ENGINE = InnoDB;
'''
```
