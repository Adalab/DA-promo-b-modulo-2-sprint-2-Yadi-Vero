```python
def insertar_tabla(nombre_bbdd, query):
    
    cnx = mysql.connector.connect(user='root', password="AlumnaAdalab",
                                     host='127.0.0.1', database=f"{nombre_bbdd}")
    mycursor = cnx.cursor()
    
    try: 
        mycursor.execute(query)
        cnx.commit() 
    except mysql.connector.Error as err:
        print(err)
        print("Error Code:", err.errno)
        print("SQLSTATE", err.sqlstate)
        print("Message", err.msg)
```

```python
tabla_ataques= '''
CREATE TABLE IF NOT EXISTS `tiburones`.`ataques` (
  `id_ataques` INT NOT NULL AUTO_INCREMENT,
  `case_number` DATE NOT NULL,
  `year` FLOAT NOT NULL, 
  `type` VARCHAR(55) NOT NULL,
  `country` VARCHAR(25) NOT NULL,
  `area` VARCHAR(25) NOT NULL,
  `location` VARCHAR(150) NOT NULL,
  `activity` VARCHAR(50) NOT NULL,
  `name` VARCHAR(150) NOT NULL,
  `age` INT NOT NULL,
  `time` VARCHAR(25) NOT NULL,
  `species_` VARCHAR(50) NOT NULL,
  `injury` VARCHAR(500) NOT NULL,
  `fecha_limpia` VARCHAR(25) NOT NULL,
  `fatal` VARCHAR(10) NOT NULL,
  `sex` VARCHAR(10) NOT NULL,
  PRIMARY KEY (`id_ataques`))
ENGINE = InnoDB;
'''
```

```python
# Debemos terminar el groupby del pair III

tabla_clima= '''
CREATE TABLE IF NOT EXISTS `tiburones`.`clima` (
  `id_clima` INT NOT NULL AUTO_INCREMENT,
  `id_ataques` INT NOT NULL,
  `pais` VARCHAR(25) NOT NULL,
  `index` FLOAT NOT NULL,
  `timepoFLOAT` FLOAT NOT NULL, 
  `cloudcover` FLOAT NOT NULL,
  `highcloud`FLOAT NOT NULL,
  `lowcloud`FLOAT NOT NULL,
  `temp2m` FLOAT NOT NULL,
  `lifted_index` FLOAT NOT NULL,
  `rh2m` FLOAT NOT NULL,
  `msl_pressure` FLOAT NOT NULL,
  `prec_amount` FLOAT NOT NULL,
  `snow_depth` FLOAT NOT NULL,
  `wind10.direction` FLOAT NOT NULL,
  `wind10M.speed` FLOAT NOT NULL,
  PRIMARY KEY (`id_clima`),
  CONSTRAINT `fk_tabla_clima_tabla_ataques`
    FOREIGN KEY (`id_ataques`)
    REFERENCES `tiburones`.`ataques` (`id_ataques`))
ENGINE = InnoDB;
'''
```
